```python
## Written by Jason Chen, former web assistant
## If you run into any questions, feel free to email me
## scierc@uw.edu
######
# This program scrape all the links within "research and collections" tab
# and download all the information of the sub pages 
# into a table format into Word.The doc is saved at the root.
# This program can also be easily edited into scraping other
# pages of the Burke website.
# Only feasible for the current Burke website template as of Jan.31st 2018
######
## parsing the collections page!!
## step 1:paste over the description text!!
## step 2:parse the two columns section into a table
## with img on the left column and text on the right column

import urllib2
import urllib
from docx import Document
from bs4 import BeautifulSoup
# initiation
url2 = "http://www.burkemuseum.org/research-and-collections"
html2 = urllib2.urlopen(url2).read() 
soup5 = BeautifulSoup(html2,"html.parser")
url_div = soup5.findAll("div", {"class":"views-field views-field-nothing"})
all_url = []
for i in url_div:
    a = i.find("a")
    a = 'http://www.burkemuseum.org' + a['href']
    all_url.append(a)

# step1
all_url = iter(all_url)
next(all_url)
    
for pp in all_url:
    print pp
    #find h1 and name it the doc!
    url3 = pp
    html = urllib2.urlopen(url3).read()
    soup3 = BeautifulSoup(html,"html.parser")
    document = Document()
    print 'initiation completed'
    #get the title of the page for the doc
    page_title = soup3.find("h1").text
    print page_title
    
    document.add_heading(page_title + ' website update', 0)
    document.add_heading('please track changes!', level=2)     
    document.add_heading('instructions:', level=1)

    ins = "Please review the following text and update any collections numbers or other outdated information. Feel free to add new collections highlights if there are new ones you would like to appear."
    document.add_paragraph(ins)
    ##for OVERVIEW page        
    #heading
    # add the heading of the section 
    
    document.add_heading('Overview', level=1)
    print 'overview title added'
    #body text
    document.add_paragraph(url3)
    overview = soup3.findAll("div", {"class":"views-field views-field-body"})
    overview_p = overview[0].findAll("p")
    document.add_paragraph(overview_p[0].text)
    print page_title + ' overview page done'
    
    ##for COLLECTION page####### soup 
    url = pp + "/collections"
    
    html = urllib2.urlopen(url).read()
    soup = BeautifulSoup(html,"html.parser")
    
    # add the heading of the section 
    intro = soup.findAll("div", {"class":"views-field views-field-title"})
    for i in intro:
        title = i.findAll("h2")
        if len(title) > 0:
            title = title[0].text
            document.add_heading(title, level=1)
    document.add_paragraph(url)
    intro_content = soup.findAll("div", {"class":"views-field views-field-field-collections-description"})[0]
    intro_p = intro_content.findAll("p")
    for i in intro_p:
        #if there is a strong! Then add it bold in word as a new paragraph
        #else this is a body text, so simply paste it over
        if len(i.findAll("strong")) > 0:
            bold_text = i.find("strong").text
            p = document.add_paragraph('')
            p.add_run(bold_text).bold = True
            body_text = i.text[len(bold_text)+1:]
            document.add_paragraph(body_text)
        else:
            document.add_paragraph(i.text)
    print 'done intro!'


    # step2
    data2 = soup.findAll("div", {"class":"view view-divisions view-id-divisions view-display-id-attachment_17"})
    data3 = data2[0].findAll("div", {"class" : lambda L: L and L.startswith('views-row views-row-')})
    count = 0
    table = document.add_table(rows=1, cols=2) 
    table.style = 'TableGrid'
    while (count < len(data3)):
        img = data3[count].find('img')
        src = str(img['src'])
        if src.startswith('http'):
            src = src
        else:
            src = 'http:' + src
        name = 'please work.jpg'
        row_cells = table.add_row().cells
        urllib.urlretrieve(src, name)
        paragraph = row_cells[0].paragraphs[0]
        run = paragraph.add_run()
        run.add_picture(name, width = 1400000, height = 1400000)

        ##inserting text in row_cells[1].text
        all_text = data3[count].findAll("div", { "class":"views-field views-field-body" })
        ##copy ALL the text
        ##part 1: paste over the text from p's
        ##part 2: if there are lists, paste over the list elements

        #part 1 #problem for Mammalogy
#         p = all_text[0].findAll("p")
#         print 'getting recursing...'                  
        for child in all_text[0].recursiveChildGenerator():
            name = getattr(child, "name", None)
            if str(name) == 'p':
                row_cells[1].add_paragraph(child.text)
            if str(name) == 'ul':
                row_cells[1].add_paragraph(child.text, style='ListBullet')
        count = count + 1
        print count
    print 'done column!' 
    document.save(page_title + ' website update' + '.docx')   
    print 'all done!'
```
